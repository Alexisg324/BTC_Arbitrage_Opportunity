```python
# For the date early in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one
arbitrage_spread_early_jan = bitstamp_sliced.loc["2018-01-15"] - coinbase_sliced.loc["2018-01-15"]

# Use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame

arbitrage_spread_early_jan[arbitrage_spread_early_jan>0].describe()
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).
spread_return_early_jan = arbitrage_spread_early_jan[arbitrage_spread_early_jan>0] / coinbase_sliced.loc["2018-01-15"]

# Review the spread return DataFrame
spread_return_early_jan.describe()
```

```python
# Middle dataset arb
spread_return_middle_feb = arbitrage_spread_middle_feb[arbitrage_spread_middle_feb>0] / coinbase_sliced.loc["2018-02-14"]

# Review the spread return DataFrame
spread_return_middle_feb.describe()
```

```python
# Late dataset arb
spread_return_late_mar = arbitrage_spread_late_mar[arbitrage_spread_late_mar>0] / coinbase_sliced.loc["2018-03-24"]

# Review the spread return DataFrame
arbitrage_spread_late_mar.describe()
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.

```python
# For the date early in the dataset, determine the number of times your trades with positive returns 
# exceed the 1% minimum threshold (.01) that you need to cover your costs
profitable_trades_early_jan = arbitrage_spread_early_jan[arbitrage_spread_early_jan>.01] / coinbase_sliced.loc["2018-01-15"]

# Review the first five profitable trades
profitable_trades_early_jan = profitable_trades_early_jan.dropna()
profitable_trades_early_jan.head()
```

```python
# Middle Dataset 
profitable_trades_middle_feb = arbitrage_spread_middle_feb[arbitrage_spread_middle_feb>.01] / coinbase_sliced.loc["2018-02-14"]
profitable_trades_middle_feb = profitable_trades_middle_feb.dropna()
profitable_trades_middle_feb.head()
```

```python
# Late Dataset
profitable_trades_late_mar = arbitrage_spread_late_mar[arbitrage_spread_late_mar>.01] / coinbase_sliced.loc["2018-03-24"]
profitable_trades_late_mar = profitable_trades_late_mar.dropna()
profitable_trades_late_mar.head()
```
